```python
a=[[1,2],[3,4]]
b=[[5,6],[7,8]]
result=[[[a[i][j]+b[i][j]]for j in range(2)]for i in range(2)]
print("matrix addition:")
for row in result:
  print(row)
```

```python
A=[[1,2],[3,4]]
B=[[2,0],[1,2]]
result=[[0,0],[0,0]]
for i in range(2):
  for j in range(2):
    for k in range(2):
      result[i][j]+=A[i][k]*B[k][j]
print("matrix manipulation:")
for roe in result:
  print(roe)
```

```python
A=[[1,2,3],[4,5,6]]
transpose=[[[A[j][i]] for j in range(len(A))]
for i in range(len(A[0]))]
print("transpose:")
for row in transpose:
  print(row)

```
