```python
def load_data(agridata):
  crops=[]
  file=open("agridata.txt","r")
  for line in file:
    line=line.strip()

    if line:
      data=line.split(",")
      crop={
          "name":data[0],
          "season":data[1],
          "area":float(data[2]),
          "rainfall":(data[3]),
          "fertilizer":(data[4]),
          "yield":float(data[5])
          }
      crops.append(crop)
  file.close()
  return crops

def calculate_yield_per_area(crop):
    return crop["yield"]/crop["area"]

def display_crops(crops):
  print("\n crops records:")
  for crop in crops:
    ypa=calculate_yield_per_area(crop)
    print(crop["name"],"_",crop["season"],"_yield/area:",round(ypa,2))


def find_highest_and_lowest_yield(crops):
  highest=crops[0]
  lowest=crops[0]
  for crop in crops:
    if calculate_yield_per_area(crop)>calculate_yield_per_area(highest):
      highest=crop
    if calculate_yield_per_area(crop)<calculate_yield_per_area(lowest):
      lowest=crop
  return highest,lowest

def seasonal_summary(crops):
  summary={}
  for crop in crops:
    season=crop["season"]
    ypa=calculate_yield_per_area(crop)
    if season not in summary:
      summary[season]=[]
    summary[season].append(ypa)
  return summary

def write_report(agridata,crops,highest,lowest,summary):
  file=open("agridata.txt","w")
  file.write("Analysis results:\n\n")

  file.write("crop yield per unit area:\n")
  for crop in crops:

    ypa=calculate_yield_per_area(crop)
    file.write(crop["name"]+":"+str(round(ypa,2))+"\n")

  file.write("\nhighest yield:\n")
  file.write(highest["name"]+":"+str(round(calculate_yield_per_area(highest),2))+"\n")
  file.write("\nlowest yield:\n")
  file.write(lowest["name"]+":"+str(round(calculate_yield_per_area(lowest),2))+"\n")

  file.write("\nseasonal average yield:\n")
  for season in summary:

    file.write(season+":"+str(round(sum(summary[season])/len(summary[season]),2))+"\n")
  file.close()

def main():
  input_file="crops.txt"
  output_file="report.txt"
  crops=load_data(input_file)

  for crop_data in crops:
    print (crop_data)
  for crop_data in crops:
    for key,value in crop_data.items():
      print(f"{key}:{value}")
      print("-----")

  display_crops(crops)
  highest,lowest=find_highest_and_lowest_yield(crops)
  summary=seasonal_summary(crops)
  write_report(output_file,crops,highest,lowest,summary)
  print("\n analysis complete.report written to",output_file)

main()
```
