```python
scan=[[1,0,1],[0,1,0],1,0,0]
alerts=[]
for i in range(len(scan)):
  row_count=0
  for j in range(len(scan[i])):
    if scan[i][j]==1:
      row_count+=1
  if row_count>1:
    alerts.append(f"row{i}quarantine")
    print(alerts)

```

```python
nums=[1,2,3]
output=[]

for i in range(len(nums)):
  for j in range(len(nums)):
    if i==j:
      continue
    if nums[i]+nums[j]>4:
      output.append((nums[i],nums[j]))

print(output)
```

```python


def nest_func(nested):
  for sublist in nested:
    if len(sublist)>1:
      print(sum(sublist))
nest_func([[1,2],[3]])
```

```python
def process_tuple(tup):
  total=0
  i=0
  while i<len(tup):
    if tup[i]%2==0:
      total+=tup[i]
    i+=1
  return total
process_tuple((10,20,30))
```
