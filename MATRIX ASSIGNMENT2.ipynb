```python
import numpy as np
#NUTRIENT AND FERTILIZER REQUIREMENTS OF 3 CROPS
nutrient_needs=np.array([[12,60,80],[150,75,100],[100,50,70]])
fertilizer_composition=np.array([[0.20,0.10,0.15],[0.15,0.20,0.10],[0.10,0.15,0.25]])
print('nutrient requirements(kg\\ha):')
print(nutrient_needs)
print('\nfertilizer composition(%):')
print(fertilizer_composition)

#MATRIX TRANSPOSE EXAMPLE
print('\n transposed nutrient needs:')
print(nutrient_needs.T)
total_nutrients_per_crop=np.sum(nutrient_needs,axis=1)
print(f'\n total nutrients needs per crop:{total_nutrients_per_crop}kg/ha')

fertilizer_application=np.array([100,100,100])
nutrients_delivered=np.dot(fertilizer_composition,fertilizer_application)
print('\nnutrients_delivered:')
print(nutrients_delivered)

#CALCULATING DETERMINANT FOR MATRIX
det=np.linalg.det(fertilizer_composition)
print(f'\ndeterminant of fertilizer:{det:6f}')

#IDENTITY MATRIX EXAMPLE(SCALING FACTOR)
scaling=np.eye(3)
enhanced_needs=np.dot(nutrient_needs,scaling)
print('\n enhanced nutrient needs(10% increase):')
print(enhanced_needs)
```
