```python
#DATA AGGREGATION AND GROUPING
import pandas as pd
#multi-year farm data
data={
    'year':[2020,2020,2020,2021,2021,2021,2022,2022,2022,2023,2023,2023],
    'rainfall':['north','south','east','north','south','east','north','south','east','north','south','east'],
    'crop':['wheat','rice','corn','wheat','rice','corn','wheat','rice','corn','wheat','rice','corn'],
    'yield_tons':[180,228,416,195,245,430,210,250,445,225,260,455],
    'area_ha':[50,45,55,52,46,56,54,48,57,55,49,58],
    'cost_thousand':[85,95,180,90,100,190,95,105,195,100,110,200]
    }
df=pd.DataFrame(data)
print('==agricultural data set==')
print(df)

#group by rainfall
print('\n===analysis by rainfall===')
rainfall_stats=df.groupby('rainfall').agg({
    'yield_tons':'sum',
    'area_ha':'sum',
    'cost_thousand':'sum'
    })
print(rainfall_stats)

#group by crop
print('\n=== analysis by crop===')
crop_stats=df.groupby('crop').agg({
    'yield_tons':['mean','max','min'],
    'cost_thousand':'mean'
    }).round(2)
print(crop_stats)

#multiple grouping rainfall
print('\n===rainfall crop analysis===')
rainfall_crop=df.groupby(['rainfall','crop'])['yield_tons'].mean().round(2)
print(rainfall_crop)

#year over growth
print('\n ===yearly trends===')
yearly=df.groupby('year').agg({
    'yield_tons':'sum',
    'area_ha':'sum'
    })
yearly['yield_per_ha']=(yearly['yield_tons']/yearly['area_ha']).round(2)
print(yearly)

#transform :add group mean to each row
df['rainfall_avg_yield']=df.groupby('rainfall')['yield_tons'].transform('mean').round(2)
df['crop_avg_yield']=df.groupby('crop')['yield_tons'].transform('mean').round(2)
print('\n===transformed data===')
print(df)

print('\n === DATASET WITH GROUP AVERAGES===')
print(df[['year','rainfall','crop','yield_tons','rainfall_avg_yield','crop_avg_yield']].head(6))

#pivot table
print('\n===pivot table:average yield by rainfall and crop===')
pivot=df.pivot_table(values='yield_tons',index='rainfall',columns='crop',aggfunc='mean').round(2)
print(pivot)

#filter groups with high performance
print('\n===high performing rainfalls(avg yield >300)===')
high_rainfalls=df.groupby('rainfall').filter(lambda x:x['yield_tons'].mean()>300)
print(high_rainfalls[['rainfall','crop','yield_tons']].drop_duplicates('rainfall'))
```

```python
#DATA AGGREGATION AND GROUPING
import pandas as pd
#multi-year farm data
data={
     'crop':['wheat','rice','corn','wheat','rice','corn','wheat','rice','corn'],
     'rainfall':[50,45,55,52,46,56,54,48,57],
     'yield_tons':[180,228,416,195,245,430,210,250,445],
     'area_ha':[10,25,15,32,16,26,34,18,17],
     }
df=pd.DataFrame(data)
print('==agricultural data set==')
print(df)

#group by rainfall
print('\n===analysis by rainfall===')
gkvk=df.groupby('crop').agg({
    'yield_tons':'sum',
    'area_ha':'sum',
    'rainfall':'sum'
    })
print(gkvk)

#group by crop
print('\n=== analysis by crop===')
crop_stats=df.groupby('crop').agg({
    'yield_tons':['mean','max','min'],
    'rainfall':'mean'
    })
print(crop_stats)

#multiple grouping rainfall
print('\n===rainfall crop analysis===')
mandya=df.groupby(['rainfall','crop'])['yield_tons'].mean()
print(mandya)

#year over growth
print('\n ===yearly trends===')
yearly=df.groupby('crop').agg({
    'yield_tons':'sum',
    'area_ha':'sum'
    })
yearly['yield_per_ha']=(yearly['yield_tons']/yearly['area_ha'])
print(yearly)

#transform :add group mean to each row
df['crop_avg_yield']=df.groupby('crop')['yield_tons'].transform('mean')
print('\n===transformed data===')
print(df)

print('\n === DATASET WITH GROUP AVERAGES===')
print(df[['rainfall','crop','yield_tons','crop_avg_yield']])

#pivot table
print('\n===pivot table:average yield by rainfall and crop===')
pivot=df.pivot_table(values='yield_tons',index='rainfall',columns='crop',aggfunc='mean')
print(pivot)

#filter groups with high performance
print('\n===high performing rainfalls(avg yield >300)===')
high_rainfall=df.groupby('rainfall').filter(lambda x:x['yield_tons'].mean()>300)
print(high_rainfall[['rainfall','crop','yield_tons']].drop_duplicates('rainfall'))
```
