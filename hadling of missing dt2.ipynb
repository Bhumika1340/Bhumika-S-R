```python
#REMOVING MISSING VALUES
df=pd.DataFrame({
    'farm_id':['f001',None,'f003','f004','f005','f006',float('nan')],
    'crop':['wheat',None,'corn','wheat','rice','corn',float('nan')],
    'yield_tons':[180,None,float('nan'),164,245,390,float('nan')],
    'rainfall_mm':[450,820,680,420,850,710,480]
    })
print(df.ffill()) #FORWARD FILL
```

```python
#INTERPOLATE
df=pd.DataFrame({
    'farm_id':['f001',None,'f003','f004','f005','f006',float('nan')],
    'crop':['wheat',None,'corn','wheat','rice','corn',float('nan')],
    'yield_tons':[180,None,float('nan'),164,245,390,float('nan')],
    'rainfall_mm':[450,820,680,420,850,710,480]
    })
print(df.interpolate())

```

```python
#Fill missing value by drawing a straight line between known data point
df=pd.DataFrame({
    'rainfall_mm':[10,20,None,40,50]
    })
print(df.interpolate())

```

```python
#REPLACE FOR ONE VALUE
df=pd.DataFrame({
    'farm_id':['f001',None,'f003','f004','f005','f006',float('nan')],
    'crop':['wheat',None,'corn','wheat','rice','corn',float('nan')],
    'yield_tons':[180,None,float('nan'),164,245,390,float('nan')],
    'rainfall_mm':[450,820,680,420,850,710,480]
    })
# Replace all occurrences of 820 with 245 in the entire DataFrame and print the result.
print(df.replace(820,245))
```

```python
#REPLACE FOR MULTIPLE VALUES
df=pd.DataFrame({
    'farm_id':['f001',None,'f003','f004','f005','f006',float('nan')],
    'crop':['wheat',None,'corn','wheat','rice','corn',float('nan')],
    'yield_tons':[180,None,float('nan'),164,245,390,float('nan')],
    'rainfall_mm':[450,820,680,420,850,710,480]
    })
# for 2 values
print(df.replace([450,850,420],[265,485,501]))
```
