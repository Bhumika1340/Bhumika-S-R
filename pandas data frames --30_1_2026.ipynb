```python
import pandas as pd
data={
    "calories":[420,380,390],
    "duration":[50,40,45]
}
df=pd.DataFrame(data)
print(df)
```

```python
#locate Row
import pandas as pd
data={
    "calories":[420,380,390],
    "duration":[50,40,45]
}
df=pd.DataFrame(data)
print(df.loc[0])
```

```python
#locate Row
import pandas as pd
data={
    "calories":[420,380,390],
    "duration":[50,40,45]
}
df=pd.DataFrame(data)
print(df.loc[0, 'duration'])
```

```python
#named indexes
import pandas as pd
data={
    "calories":[420,380,390],
    "duration":[50,40,45]
}
df=pd.DataFrame(data,index=["day1","day2","day3"])
print(df)
```

```python
#named indexes
import pandas as pd
data={
    "calories":[420,380,390],
    "duration":[50,40,45]
}
df=pd.DataFrame(data,index=["day1","day2","day3"])
print(df.loc["day2"])
```

```python
import pandas as pd
data=[
    ["alia",20,80],
    ["babu",21,85],
    ["charu",22,90]
]
df=pd.DataFrame(data,columns=["Name","Age","Marks"],index=["a","b","c"])
print(df)
```

```python
#list of dictionaries:
import pandas as pd
data=[
    {"name":"alia","age":20,"marks":80},
    {"name":"babu","age":21,"marks":85},
    {"name":"charli","age":22,"marks":90}
]
df=pd.DataFrame(data,index=["q12","q22","q13"])
print(df)
```

```python
#dictionary of series:
import pandas as pd
S1=pd.Series([10,20,30])
S2=pd.Series([40,50,60])
df=pd.DataFrame({"col1":S1,"col2":S2})
print(df)
```
