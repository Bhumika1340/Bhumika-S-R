```python
farm=dict(
    name="ashwini",
    survey_number=9,
    village="bengaluru",
    age=40
)
print("original dictionary",farm)
farm_copy=farm.copy()
print("copied dictionary:",farm_copy)
removed_value=farm.pop("age")
print("removed value:",removed_value)
print("after pop():",farm)
last_item=farm.popitem()
print("removed item:",last_item)
print("after popitem():",farm)
farm.update({"yield":25,"village":"mandya"})
print("after update():",farm)
farm.clear()
print("after clear():",farm)
del farm
```
