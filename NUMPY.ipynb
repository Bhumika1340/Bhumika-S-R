```python
print("EXAMPLE")

import numpy as np

#crop yields of 4 farms over 5 seasons in tons
crop_yields=np.array([
    [12.5,13.2,14.1,13.8,15.2],
    [10.3,11.5,10.8,12.2,11.9],
    [15.1,14.8,16.2,15.5,16.8],
    [11.2,10.9,12.3,11.7,12.8]
    ])

print('crop yields array(tons):')
print(crop_yields)
print(f'\n Shape:{crop_yields.shape}')
print(f'data type:{crop_yields.dtype}')
print(f'total elements:{crop_yields.size}')
print(f'number of dimensions:{crop_yields.ndim}')

#AGRICULTURAL ANALYSIS OPERATIONS

total_yield=np.sum(crop_yields)
avg_yield_per_farm=np.mean(crop_yields,axis=1)
avg_yield_per_season=np.mean(crop_yields,axis=0)
max_yield=np.max(crop_yields)
min_yield=np.min(crop_yields)

print(f'\n total production:{total_yield}tons')
print(f'average per farm:{avg_yield_per_farm}tons')
print(f'average per season:{avg_yield_per_season}tons')
print(f'maximum yield:{max_yield}tons')
print(f'minimum yield:{min_yield}tons')

#find best performing farm
best_farm_index=np.argmax(avg_yield_per_farm) # Corrected: use avg_yield_per_farm to find best farm index
print(f'\n best performing farm:farm{best_farm_index+1}')
print(f'their average yield:{avg_yield_per_farm[best_farm_index]:.2f}tons')
```
