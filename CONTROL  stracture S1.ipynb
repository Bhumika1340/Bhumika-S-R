```python
print("===crop suitability check===")
soil_moisture=float(input("enter soil moisture percentage:"))
rainfall=float(input("enter rainfall in mm:"))
if soil_moisture>=70 and rainfall>=200:
  print("suitable for rice")
else:
  print("not suitable for rice")
```

```python
print("fertilizer recommendation system")
crop=input("enter crop name(rice,wheat,maize):")
nitrogen=int(input("enter soil nitrogen level:"))
if crop=="rice" and nitrogen<50:
  print("apply urea")
elif crop=="wheat" and nitrogen<40:
  print("apply DAP")
else:
  print("fertilizer not required")
```

```python
print("===crop yield classidfication===")
yield_amount = float(input("enter crop yield in quintals:"))
if yield_amount>=50:
  print("high yield")
elif 30<=yield_amount<=49:
  print("medium yield")
else:
  print("low yield")
```
