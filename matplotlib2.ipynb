```python
import matplotlib.pyplot as plt
import numpy as np
x=np.random.rand(50)
y=np.random.rand(50)
sizes=1000*np.random.rand(50)
colors=np.random.rand(50)

plt.scatter(x,y,s=sizes,c=colors,cmap='phase',alpha=0.9)
plt.colorbar()
plt.title("Advanced Scatter Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```
