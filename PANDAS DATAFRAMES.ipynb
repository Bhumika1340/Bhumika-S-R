```python
#CREATING DATAFRAMES FROM ADICTIONARY OF LISTS

import pandas as pd
data={
    "name":["alice","bob","charlie"],
    "age":[20,22,21],
    "marks":[85,80,90]
    }
df=pd.DataFrame(data)
print(df)
```

```python
#CREATING DATAFRAMES FROM LIST OF LISTS

import pandas as pd
data=[
    ["alice",20,85],
    ["bob",22,90],
    ["charlie",21,88]
    ]
df=pd.DataFrame(data,columns=["name","age","marks"])
print(df)
```

```python
#CREATING DATA fRAMES FROM LIST OF DICTIONARIES
import pandas as pd
data=[
    {"name":"alice","age":20,"marks":85},
    {"name":"bob","age":22,"marks":90}
    ]
df=pd.DataFrame(data)
print(df)
```

```python
#CREATING DATAFRAMES FROM A DICTIONARY OF SERIES

import pandas as pd
s1=pd.Series([10,20,30],
             index=["a","b","c"])
s2=pd.Series([40,50,60],
             index=["a","b","c"])
df=pd.DataFrame({"col1":s1,"col2":s2})
print(df)
```
