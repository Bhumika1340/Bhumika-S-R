```python
#STEP 1: IMPORT LIBRARIES
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,r2_score

#STEP 2: CREATE SAMPLE AGRICULTURAL DATASET
print("AGRICULTURAL DATA SET")
data={
    "rainfall":[450,520,390,610,480,530,410,590],
    "temperature":[22,20,25,19,23,21,24,18],
    "nitrogen":[0.8,1.0,0.6,1.2,0.9,1.1,0.7,1.3],
    "fertilizer":[120,140,100,160,130,150,110,170],
    "yield":[3.5,4.2,2.9,4.8,3.8,4.4,3.1,5.0]
    }
df=pd.DataFrame(data)
print(df)

#STEP 3:DEFINE FEATURES (X) AND TARGET (Y)
print("\nFEATURES AND TARGET")
x=df[["rainfall","temperature","nitrogen","fertilizer"]]
y=df["yield"]
print(x)
print(y)

#STEP 4:SPLIT DATA INTO TRAINING AND TESTING SETS
print("\nTRAINING AND TESTING SETS")
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
print(x_train)
print(x_test)
print(y_train)
print(y_test)

#STEP 5:TRAIN MODEL
print("\nMODEL TRAINING")
model=LinearRegression()
model.fit(x_train,y_train)

#STEP 6:MAKE PREDICTIONS
print("\nMAKING PREDICTIONS")
y_pred=model.predict(x_test)
print(y_pred)
x_pred=model.predict(x_train)
print(x_pred)

#STEP 7:EVALUATE MODEL
print("\nMODEL EVALUATION")
mae=mean_absolute_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
print("Mean Absolute Error:",round(mae,2))
print("R2Score:",round(r2,2))

#STEP 8:PREDICT YIELD FOR NEW FARM CONDITIONS
print("\nPREDICTING YIELD FOR NEW FARM CONDITIONS")
new_data=np.array([[500,21,1.0,145]]) # Rainfall ,Temperature,nitrogen,fertilizer
predicted_yield=model.predict(new_data)
print("Predicted Yield(tons/hectare):",round(predicted_yield[0],2))


```
