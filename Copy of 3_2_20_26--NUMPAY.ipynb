```python
# agriculture application
import numpy as np
crop_yields=np.array([
    [12.3,13.5,14.5,15.5],
    [10.1,10.2,10.3,10.4],
    [15.1,15.2,15.3,15.4],
    [11.2,11.3,11.4,11.5]

])

print('crops yields array(tons):')
print(crop_yields)
print(f'\nshape:{crop_yields.shape}')
print(f'Data type:{crop_yields.dtype}')
print(f'Total elements:{crop_yields.size}')
print(f'Number of dimensions:{crop_yields.ndim}')

total_production = np.sum(crop_yields)
avg_yield_per_farm = np.mean(crop_yields,axis=1)
avg_yield_per_season = np.mean(crop_yields,axis=0)
max_yield = np.max(crop_yields)
min_yield = np.min(crop_yields)

print(total_production)
print(avg_yield_per_farm)
print(avg_yield_per_season)
print(max_yield)
print(min_yield)
print(f'\nTotal production:{total_production}tons')
print(f'averege per farm={avg_yield_per_farm}tons')
print(f'averege per season={avg_yield_per_season}tons')
print(f'maxium yield={max_yield}tons')
print(f'minimum yield={min_yield}tons')
```

```python
import numpy as np

#locality of reference

my_array = np.arange(100)
total_sum = 0
for i in range(100):
    total_sum += my_array[i]

print(f"Total sum: {total_sum}")
```
