```python
def find_max_rainfall(crops_list):
  if not crops_list:
    return 0

  max_rain = crops_list[0]["rainfall"]
  for crop_item in crops_list:
    if crop_item["rainfall"] > max_rain:
      max_rain = crop_item["rainfall"]
  return max_rain

def rainfall_efficiency_analysis(crops_list, max_rain_threshold):
  print(f"Crops with rainfall <= {max_rain_threshold}:")

  for crop_item in crops_list:
    if crop_item["rainfall"] <= max_rain_threshold:
      print(f"{crop_item['name']} - {crop_item['yield']}")

crops_data = [
    {"name": "Wheat", "yield": 2500,
     "fertilizer":150,"rainfall":800},
    {"name": "Rice", "yield": 3000,
     "fertilizer":200,"rainfall":1200},
    {"name": "Corn", "yield": 800,
     "fertilizer":100,"rainfall":500},
    {"name": "Barley", "yield":1500,
     "fertilizer":120,"rainfall":200},
    {"name": "Soybean", "yield": 1724,
     "fertilizer":180,"rainfall":150}
]


max_rainfall = find_max_rainfall(crops_data)
rainfall_efficiency_analysis(crops_data, max_rainfall)
```
