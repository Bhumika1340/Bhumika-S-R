```python
import numpy as np
nutrient_needs=np.array(
    [
    [128,60,80],
    [158,75,100],
    [100,50,70]
])
fertilizer_composition=np.array(
    [
    [0.20,0.10,0.15],
    [0.15,0.20,0.10],
    [0.10,0.15,0.75]
])
print('nutrient requierments(kg/ha):')
print(nutrient_needs)
print('/fertilizer composition(%):')
print(fertilizer_composition)

total_nutrients_per_crop=np.sum(nutrient_needs,axis=1)
print(f'\nTotal nutrients needed per crop:{total_nutrients_per_crop}kg/ha')

fertilizer_application=np.array([100,180,100])#kg of each type
nutrients_delivered=np.dot(fertilizer_application,fertilizer_composition)
print(f'\nNutrients from 100kg of each type of fertilizer:{nutrients_delivered}')

det=np.linalg.det(fertilizer_composition)
print(f'\nDeterminant of fertilizer composition matrix:{det:6f}')

scaling=np.eye(3) *1.1 #10%increases
enhanced_needs=np.dot(nutrient_needs,scaling)
print('\nEnhanced nutrient needs(10% increase):')
print(enhanced_needs)
```
