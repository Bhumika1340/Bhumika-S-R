```python
number_of_fields=5
area_per_field=2.5
crop_name="rice"
irrigation_available=True
print("crop name:",crop_name)
print("number of fields:",number_of_fields)
print("Area per field:",area_per_field,"acres")
print("Irrigation available:",irrigation_available)
total_area=number_of_fields*area_per_field
print("\n total cultivated area:",total_area,"acres")
if irrigation_available:
  print("status: irrigation facility available")
else:
  print("status: no irrigation facility")
```
