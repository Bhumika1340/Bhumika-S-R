```python
codes=[
    "CRP_RICE_2026",
    "CRP_WHEAT_2025",
    "CRP_MAIZE_2024",
    "FRT_MANGO_2026"
    ]
for code in codes:
  if code.startswith("CRP_"):
    code,crop,year=code.split("_")
    print(crop,year)
```

```python
raw_record="rice,1200,black soil"
clean_record=raw_record.strip()
clean_record1=clean_record.rstrip()
crop,yield_kg,soil=clean_record1.split(",")
crop=crop.title()
print(f"crop:{crop}")
print(f"yield:{yield_kg}kg")
print(f"soil:{soil}")
```
