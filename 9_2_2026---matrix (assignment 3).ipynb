```python
import numpy as np
yields=np.array([
    [45,48,52,49,53],
    [38,41,39,44,42],
    [52,55,58,56,60],
    [41,43,45,47,46]
])

weather_factors=np.array([1.0,0.95,1.05,0.98,1.02])
print('original yields(tons/hectare):')
print(yields)
print('\nWeather Impact Factors:',weather_factors)

#broadcasting:
adjusted_yields=yields*weather_factors
print('\nAdjusted Yields(tons/hectare):')
print(adjusted_yields.round(2))

#boolean indexing:
exceptional=adjusted_yields>55
print('\nExceptional Yields(>55):')
print(adjusted_yields[exceptional])

#conditional operations:
categories=np.where(
    adjusted_yields>50,'High',
    np.where(adjusted_yields>40,'Medium','Low')
)
print('\nYield Categories:')
print(categories)

#Advanced indexing:
selected_farms=[0,2] #Farm1 and Farm3
print(f'\nSelected farms yields:')
print(yields[selected_farms])

#Multi-condition filtering:
good_years=adjusted_yields[(adjusted_yields>45)&(adjusted_yields<55)]
print('\nGood performing Years(45-55tons):')
print(good_years)
```

```python
import numpy as np
a=np.arange(50).reshape(10,5)
print("original array(10*5):\n",a)

#select all rows where the first festure
# (column 0) is greater than 20
print(a[:])
print(a[:,0])
print(a[:,1])
selected_rows=a[a[:,0]>20]
print("\nRows where first festure>20:\n",selected_rows)
```
