```python
from random import choices
def add(x,y):
 return x+y

def sub(x,y):
 return x-y

def multiply(x,y):
 return x*y

def divide(x,y):
 if y==0:
  return "error!division by zero"
 else:
  return x/y
def get_input():
  x=float(input("enter the first number"))
  y=float(input("enter the second number"))
  return x,y

def print_menu():
  print("\n select operation:")
  print("1.addition")
  print("2.subtraction")
  print("3.multiplication")
  print("4.division")
  print("5.exit")

def calculator():
  while True:
    print_menu()

    choices=input("enter your choice(1/2/3/4/5):")
    if choices=="5":
      print("exiting calculator")
      break
    x,y=get_input()
    if choices=="1":
      result=add(x,y)
    elif choices=="2":
      result=sub(x,y)
    elif choices=="3":
      result=multiply(x,y)
    elif choices=="4":
      result=divide(x,y)
    else:
      print("invalid choice")
      continue
    print("result:",result)
calculator()
print("done with the programme")











```
